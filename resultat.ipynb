```python
from bs4 import BeautifulSoup
from passport import Passport
import re
import os
```

```python
_inputfile = './ressource/converted-text/'
_person = list()

for file in os.listdir(_inputfile):

    hocr = open(_inputfile + file, 'r').read()
    soup = BeautifulSoup(hocr, 'html.parser')
    html = list(soup.children)[4]
    body = list(html.children)[3]
    div = list(body.children)[1]
    
    text = div.find_all('span', class_='ocr_line')
    for loop in text:
        print(loop.get_text())
            
    text = div.find_all('span', class_='ocr_line')
   
    nom = ""
    prenom = ""
    nationalite = ""
    num_passport = ""
    date_naissance = ""
    date_exp = ""
    mrz_1 = ""
    mrz_2 = ""
    
    for loop in text:
                
        if re.search(r"Nom", loop.get_text()) is not None or re.search(r"NOM /", loop.get_text()) is not None:
            print(loop.get_text())
            print(loop.find_next('span', class_='ocr_line').get_text())
            nom = str(loop.find_next('span', class_='ocr_line').get_text() + "").split("\n")
            nom = nom[0].split(" ")
            nom = nom[0].split(",")
            if len(nom) > 1:
                nom = nom[1]
            else:
                nom = nom[0]
        
        if re.search(r"Post", loop.get_text()) is not None or re.search(r"Postn", loop.get_text()) is not None:
            print(loop.get_text())
            print(loop.find_next('span', class_='ocr_line').get_text())
         #   nom = nom  + str(loop.find_next('span', class_='ocr_line').get_text() + "")

        if re.search(r"Given", loop.get_text()) is not None or re.search(r"GIVEN", loop.get_text()) is not None or re.search(r"Prénom", loop.get_text()) is not None:
            print(loop.get_text())
            print(loop.find_next('span', class_='ocr_line').get_text())
            prenom = str(loop.find_next('span', class_='ocr_line').get_text() + "").split("\n")
            prenom = prenom[0].split(" ")
            prenom = prenom[0]

        if re.search(r"Natio", loop.get_text()) is not None or re.search(r"NATION", loop.get_text()) is not None or re.search(r"Natonaité", loop.get_text()):
            print(loop.get_text())
            print(loop.find_next('span', class_='ocr_line').get_text())
            nationalite = str(loop.find_next('span', class_='ocr_line').get_text()).split("\n")
            nationalite = nationalite[0].split(" ")
            nationalite = nationalite[0]

        if re.search(r"Type", loop.getText()) is not None or re.search(r"TYP", loop.get_text()) is not None:
            print(loop.get_text())
            print(loop.find_next('span', class_='ocr_line').get_text())
            num_passport = str(loop.find_next('span', class_='ocr_line').get_text()).split("\n")
            num_passport = num_passport[0]

        if re.search(r"Date de nas sence", loop.get_text()) is not None or re.search(r"Date de naissance", loop.get_text()) is not None or re.search(r"Da*ce", loop.get_text()) is not None or re.search(r"DATE DE NAISSANCE", loop.get_text()) or re.search(r"Daie de oaiesancés", loop.get_text()) is not None:
            print(loop.get_text())
            print(loop.find_next('span', class_='ocr_line').get_text())
            date_naissance = str(loop.find_next('span', class_='ocr_line').get_text() + "").split("\n")
            date_naissance = date_naissance[0].split(" ")
            if len(date_naissance) > 3:
                if date_naissance[3] == "ETUDLANT":
                    date_naissance = date_naissance[0] + " " + date_naissance[1] + " "+ date_naissance[2]
                if date_naissance[3] == "M":
                    date_naissance = "23 " + date_naissance[1] + " 19" + date_naissance[2]
            if date_naissance[0] == "<":
                date_naissance = date_naissance[1]
            if len(date_naissance) == 5:
                date_naissance = date_naissance[1] + " " + date_naissance[2] + " " + date_naissance[3]
            if len(date_naissance) == 1 or len(date_naissance) == 2:
                date_naissance = date_naissance[0]
            if nationalite == "CAMEROONIAN":
                date_naissance = date_naissance[0:18]

        if re.search(r"expiration", loop.get_text()) is not None or re.search(r"expi", loop.get_text()) is not None:
            print(loop.get_text())
            print(loop.find_next('span', class_='ocr_line').get_text())
            date_exp = str(loop.find_next('span', class_='ocr_line').get_text() + "").split("\n")
            date_exp = date_exp[0]
            
        if re.search(r"de<d$", loop.get_text()) is not None or re.search(r"<<$", loop.get_text()) is not None or re.search(r"x<$", loop.get_text()) is not None or re.search(r"s<$", loop.get_text()) is not None:
            mrz_1 = loop.get_text()
            mrz_2 = loop.find_next('span', class_='ocr_line').get_text()
            mrz_11 = mrz_1.split('<')
            mrz_22 = mrz_2.split('<')
            print(mrz_1)
            print(mrz_11)
            
            if len(nom) == 0:
                if len(mrz_11[0]) > 1:
                    nom = mrz_11[1]
                else:
                    nom = mrz_11[2]
            if len(prenom) == 0:
                prenom = mrz_11[4]
                if len(mrz_11[5]) > 1:
                    prenom = prenom + " " +mrz_11[5]
                if nationalite == "":
                    if len(mrz_11[0]) == 1:
                        nationalite = mrz_11[1]
                        nationalite = nationalite[0:3]
                    else:
                        nationalite = mrz_11[0]
                        nationalite = nationalite[0:3]
            print(mrz_2)
            print(mrz_22)
            if len(num_passport) > 9 or len(num_passport) < 6 or re.search(r"[0-9]$", num_passport) is None:
                num_passport = mrz_22[0]
                num_passport = num_passport[0:9]
            if re.search(r"[0-9]$", date_exp) is None:
                date_exp = date_exp.split(" ")
                date_exp = date_exp[0]
            if len(date_exp) < 5:
                mrz_222 = mrz_2[15:30]
                mrz_222 = mrz_222.split("M")
                mrz_222 = mrz_222[1]
                aa = mrz_222[0:2]
                mm = mrz_222[2:4]
                jj = mrz_222[4:6]
                date_exp = str(jj) + "/"+ str(mm) + "/20" + str(aa)
            if len(date_naissance) == 0:
                mrz_2222 = mrz_2.split(nationalite)
                mrz_2222 = mrz_2222[1]
                aa = mrz_2222[0:2]
                mm = mrz_2222[2:4]
                jj = mrz_2222[4:6]
                date_naissance = str(jj) + "/" + str(mm) + "/19" + str(aa)
        
    _person.append(Passport(nom,prenom,nationalite,num_passport,date_naissance,date_exp))
```

```python
import pandas as pd

noms = [nom.nom for nom in _person]
prenoms = [prenom.prenom for prenom in _person]
nationalites = [nationalite.nationalite for nationalite in _person]
num_passports = [num_passport.num_passport for num_passport in _person]
date_naissances = [date_naissance.date_naissance for date_naissance in _person]
date_exps = [date_exp.date_exp for date_exp in _person]
    
personne = pd.DataFrame({
"Nom": noms,
"Prenom": prenoms,
"Nationalité": nationalites,
"N° Passport": num_passports,
"Date Naissance": date_naissances,
"Date Expiration": date_exps
})
personne
```
